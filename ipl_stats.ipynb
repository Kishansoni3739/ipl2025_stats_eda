## Most Numbers of Wins

```python
wins = df25['winner'].value_counts().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=wins.values, y=wins.index)

plt.xlabel("Wins")
plt.ylabel("Team")
plt.title("Wins by Team (Season 18)")
plt.show()
```
