```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("ipl.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
df25 = data[data['season'] == 18]
```

```python
df25.head()
```

```python
df25 = df25.reset_index()
```

```python
df25.head()
```

```python
df25.columns
```

```python
# Maximum Times Toss Winner
df25.groupby('toss_winner').size().idxmax()
```

## Toss Decision By Team

```python
toss_table = df25.groupby(['toss_winner', 'toss_decision']).size().unstack(fill_value=0)
toss_table
```

```python
toss_table.plot(kind='bar', stacked=True)
plt.ylabel("Matches")
plt.title("Toss Decision by Team")
plt.show()
```

## Most Numbers of Wins

```python
wins = df25['winner'].value_counts().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=wins.values, y=wins.index)

plt.xlabel("Wins")
plt.ylabel("Team")
plt.title("Wins by Team (Season 18)")
plt.show()
```

```python
df25.head()
```

## Most Player of the Match - Award

```python
most_pom = (
    df25.groupby('player_of_match')
        .size()
        .reset_index(name='pom_count')
        .sort_values(by='pom_count', ascending=False)
)

most_pom.head(20)
```

```python
sns.barplot(x = "player_of_match", y = "pom_count", data = most_pom.head(20), palette='viridis')
plt.xticks(rotation = 90)
plt.xlabel("Player")
plt.ylabel("POM Awards")
```

## Venue wise Average Team 1 Score

```python
target_score = df25.groupby('venue')['target_runs'].mean().reset_index()
```

```python
target_score.columns = ['Venue', 'Average_Score']
```

```python
target_score
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=target_score,
    x='Average_Score',
    y='Venue',
    palette='viridis'
)

plt.title("Average Target Score by Venue (Season 18)")
plt.xlabel("Average Score")
plt.ylabel("Venue")
plt.show()
```

## Venue wise number of matches

```python
venue_matches = df25.groupby('venue')['index'].count().reset_index()
```

```python
venue_matches.columns = ['venue', 'number of matches']
```

```python
sns.barplot(x=venue_matches['venue'].str.split(' ').str[-1], y ='number of matches', data = venue_matches)
plt.xticks(rotation =90)
plt.xlabel('Venue')
plt.ylabel('Number Of Matches')
plt.title('Venue wise Number of Matches')
plt.show()
```
